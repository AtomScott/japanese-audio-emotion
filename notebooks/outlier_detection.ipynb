# Before outlier detection

```python
import pathlib

dir = pathlib.Path("./emotion/datasets/XperFace/images/embeddings")
paths = list(dir.glob("**/*.npy"))
print(len(paths))
```

```python
import numpy as np

embeddings = []
for path in paths:
    embeddings.append(np.load(path))

embeddings = np.asarray(embeddings)
print(embeddings.shape) 
```

```python
import numpy as np
from sklearn.manifold import TSNE

embeddings_2d = TSNE(n_components=2).fit_transform(embeddings)
embeddings_2d.shape

```

```python
import matplotlib.pyplot as plt

plt.scatter(*embeddings_2d.T)
plt.show()
```

# After outlier detection

```python
dir = pathlib.Path("./emotion/datasets/XperFace/images/purified_embeddings")
paths = list(dir.glob("**/*.npy"))
print(len(paths))

embeddings = []
for path in paths:
    embeddings.append(np.load(path))

embeddings = np.asarray(embeddings)
print(embeddings.shape) 

embeddings_2d = TSNE(n_components=2).fit_transform(embeddings)
embeddings_2d.shape

plt.scatter(*embeddings_2d.T)
plt.show()
```
