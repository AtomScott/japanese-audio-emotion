```python
from pathlib import Path
from JAVER.track_tools.face_tracker import FaceTracker

# setup face tracker
p = Path("../../tests/data/Elon Musk/")
ref_paths = list(p.glob("inliers/*"))

face_tracker = FaceTracker(
    image_size=160, 
    ref_paths=ref_paths, 
    batch_size=8, 
    step_large=150, 
    step_small=10
)
```

```python
video_path = './20_sample_in/single_person.mp4'
out_dir = './20_sample_out/single_person'
face_tracker.track([video_path], out_dir)
```
